Démonstration de l'API pour la phase ml

```python
import requests
import pandas as pd
import json

BASE_URL = 'http://localhost:8000'
print('✅ Prêt à utiliser l\'API')
```

## Générer un dataset pour ml

```python
response = requests.post(
    f'{BASE_URL}/dataset/generate',
    json={
        'phase': 'ml',
        'seed': 42,
        'n': 1000
    }
)
data = response.json()
dataset_id = data['meta']['dataset_id']
print(f'Dataset ID: {dataset_id}')
df = pd.DataFrame(data['result']['data_sample'])
df.head()
```
